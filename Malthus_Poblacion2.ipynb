### Modelo de Malthus con Python - 02
# Dinámica de poblaciones
#### Solucion General
$$ P(t) = Ce^{kt} $$

En un cultivo de bacterias, se estimó que inicialmente había 150 bacterias y 200 después de una hora.
¿Cuántas bacterias habría en 2 horas ?
#### Solucion General
                 $$ P(t) = Ce^{kt} $$
#### Solucion General
                  $$ P(t) = 150e^{Ln(4/3)} $$
#### Condiciones
 $$ t = 0, P = 150 $$
 $$ t = 1, P = 200 $$
 $$ t = 2, P = ? $$
 $$ P(2) = 150e^{Ln(4/3)} $$

#### solución
$$P(2) = 266.66 $$

```python
%matplotlib inline

# librerias
import matplotlib.pyplot as plt
import numpy as np
import sympy 
import pandas as pd


from scipy import integrate
from math import log

# imprimir con notación matemática.
sympy.init_printing(use_latex='mathjax')
```

```python
# generar los valores para t
t = np.arange(0, 5.5, 0.5)
#t
#s=np.exp(log(2)*t)
#s

#e=np.exp(1)
#e

# solución a la Ecuación Diferencial
# generar el valor de y de la Ec. diferencial
s = 150 * (np.exp((log(4/3) * t)))

#s
```

```python
# ver los datos
d=(t,s)

df = pd.DataFrame(d)
df
```
